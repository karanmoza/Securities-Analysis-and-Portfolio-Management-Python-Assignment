```python
#Install required libraries
!pip install yfinance
import matplotlib.pyplot as plt
import statsmodels.api as sm
import yfinance as yf
import pandas as pd
import numpy as np
import seaborn as sns
import scipy.stats as stats
import warnings
warnings.filterwarnings("ignore")
```

```python
TVS_Motorsdf = yf.download('TVSMOTOR.NS', 
                      start='2015-12-16', 
                      end='2017-01-04', 
                      progress=False
)
TVS_Motorsdf.head()
```

```python
nifty_df = yf.download('^NSEI',start='2015-12-16', 
                      end='2017-01-04', 
                      progress=False
)
nifty_df.head()
```

```python
#Calculate Log returns
nifty_df['Return_NIFTY'] = np.log(nifty_df['Adj Close']/nifty_df['Adj Close'].shift(1))
print(nifty_df['Return_NIFTY']) 
```

```python
#removing one enties from NIFTY 50 data
nifty_df.dropna(inplace=True)
```

```python
TVS_Motorsdf['Return_TVS_Motors'] = np.log(TVS_Motorsdf['Adj Close']/TVS_Motorsdf['Adj Close'].shift(1))
print(TVS_Motorsdf['Return_TVS_Motors'])
```

```python
#removing one enties from company data
TVS_Motorsdf.dropna(inplace=True)
```

```python
TVS_Motorsdf
```

```python
TVS_Motorsreturns = TVS_Motorsdf['Return_TVS_Motors']
TVS_Motorsreturns.head()
```

```python
niftyreturns = nifty_df['Return_NIFTY']
niftyreturns.head()
```

```python
#Combining the NIFTY 50 and Company returns 
returns = pd.merge(niftyreturns,TVS_Motorsreturns,how='inner',on='Date')
returns.head()
```

Regression by Market Model

```python
x = returns['Return_NIFTY']
y = returns['Return_TVS_Motors']
x_sm = sm.add_constant(x)
```

```python
#Regression - Market model 
marketmodel = sm.OLS(y,x_sm)
```

```python
resultsmm = marketmodel.fit()
```

```python
resultsmm.summary()
```

```python
betaTVS_Motorsmm=resultsmm.params[1]
```

```python
resultsmm.f_pvalue
```

```python
returnsTVS_Motors=returns
```

```python
returnsTVS_Motors['ER_MM']=returnsTVS_Motors['Return_NIFTY']*betaTVS_Motorsmm
```

```python
returnsTVS_Motors
```

Using CAPM

```python
rfr=0.02/100
```

```python
returnsTVS_Motors['Rm-Rf']=returnsTVS_Motors['Return_NIFTY']-rfr
```

```python
x = returns['Rm-Rf']
y = returns['Return_TVS_Motors']
x_sm = sm.add_constant(x)
```

```python
capmmodel = sm.OLS(y,x_sm)
```

```python
resultscapm = capmmodel.fit()
```

```python
resultscapm.summary()
```

```python
returnsTVS_Motors['ER_CAPM']=rfr+resultscapm.params[1]*returnsTVS_Motors['Rm-Rf']
```

Description of various factors used in above APM

-
- Market Model:-In the CAPM model, we use the risk premium as the independent variable and the actual returns of the security as the dependent variable for the regression.

- CAPM:-For the market index model, we use the actual return of the index as the independent variable and the actual returns of the security as the dependent variable for the regression.

- FF3:-For Fama 3 factor model, we defined big firms (B) as the top 10% by market capitalization and classified the remaining firms as small firms (S), Growth stocks (low value group), G, comprised the bottom 30% stocks in terms of the B/M ratio.The remaining stocks were grouped as neutral (N) stocks, the value factor HML (High minus Low) is defined as the simple average of two differences: (Small Value companies - Small Growth companies) and (Big value companies - Big growth companies), the value factor SMB (Small minus Big) is the simple average of three return differences: (Small Growth companies − Big growth companies), (Small neutral companies − Big Neutral companies) and (Small Value companies − Big value companies) High value group, V , consisted of the top 30% stocks in terms of the B/M ratio. The third factor being the market risk premium to compute beta via regression.


- FF4:-For Fama 4 factor model, we used HML, SMB and market risk premium as in the 3 factor model and in additon, the momentum factor MF for the regression to find beta.
